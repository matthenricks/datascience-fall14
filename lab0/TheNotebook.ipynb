## Creating Beautful Plots/Sending Out Data

In this section, we will import in all the data, generating plots for it for each of the analyses types.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# For the sole purpose of increase readability
pd.options.display.mpl_style = 'default'
# Render our plots inline
%matplotlib inline

# For use in error calculation
import scipy.stats as ss 
```

```python
# You need to always re-run the full analysis unless you add another sampling method

sourcesToRead = ['EnronAll/analysis/', 'Twitter-Ego5/analysis/', 'EpinionsAnalysis/analysis/', 'ErdosAllNEW/analysis/']
graphs = ['Enron', 'Twitter-Ego', 'Epinions', 'Erdos']
filesToRead = ['RNDanalysis.csv','RNDWalkanalysis.csv','MRNDWalkanalysis.csv', 'BFSRNDanalysis.csv', 'OLDBFSRNDanalysis.csv']
print sourcesToRead
print graphs
print filesToRead
# BFS:RND needs to be FF:RND --> 'ForestFireanalysis.csv'

alpha = [1,2,3,4,5,8,10]
samplingMethods = ['Random','Random Walk','MH Random Walk', 'BFS', 'Old BFS']
measuresTaken = ['InDegree Distribution', 'OutDegree Distribution', 'Ego-Centric Density', 'Betweenness Centrality']
```

### Data Export Start

```python
#Only doing Precision in this section
columns = ['Parent Node Count', 'Parent Edge Count', 'Replica ID', 'Alpha (%)', 'Measure', 'Population P/R Alpha', 'Sample P/R Alpha', 'Precision', 'Recall']

print '\033[5m' + 'Precision-Recall Printout' + '\033[0m'
dataBulk = pd.DataFrame(columns = ['Graph'] + columns[0:2] + ['Sampling'] + columns[2:])
z = 0
for fileHeader in sourcesToRead:
    i = 0
    for samplingMethod in filesToRead:
        csvFile =  pd.read_csv(fileHeader + samplingMethod, header=0, 
        converters={'Alpha (%)': lambda x: float(x.replace('%', '')),
                    'BFS:RND': lambda x: float(x.replace('%', '')),
                    'Percentage Amount': lambda x: float(x.replace('%', '')),
                    'Sample Correlation Alpha (%)': lambda x: float(x.replace('%', '')),
                    'Population P/R Alpha': lambda x: float(x.replace('%', ''))
                    })
        # Remove duplicates
        theFrame = pd.DataFrame(data=csvFile[columns])
        theFrame.drop_duplicates(cols=columns, take_last=True, inplace = True)
        theFrame['Sampling'] = samplingMethods[i]
        theFrame['Graph'] = graphs[z]
        dataBulk = pd.concat([dataBulk, theFrame])
        i = i + 1
    z = z + 1
# Print out
dataBulk.to_csv("files/PR.csv", header = True)
print 'DONE'
```

### END DATA EXPORT

```python
#Only doing Precision in this section
columns = ['Parent Node Count', 'Parent Edge Count', 'Sample Method Type', 'BFS:RND', 'Replica ID', 'Alpha (%)', 'Sample Node Count', 'Sample Edge Count', 'Measure', 'Population P/R Alpha', 'Sample P/R Alpha', 'Precision', 'Recall']
grouping = ['Parent Node Count', 'Parent Edge Count', 'Sample Method Type', 'BFS:RND', 'Alpha (%)', 'Measure', 'Population P/R Alpha', 'Sample P/R Alpha']

precisionOptions = [0.1, 1.0]

print '\033[5m' + 'Precision' + '\033[0m'

for measure in measuresTaken:        
    # Make it known what the measure is
    print measure
    i = 1
    fig = plt.figure(figsize=(30, 10))
    for precision in precisionOptions:
        # Precision is known as part of the y axis
        for samplingMethod in filesToRead:
            # Add the plot and the sampling method corresponding
            plot = fig.add_subplot(len(precisionOptions), len(samplingMethods), i)
            plot.set_title(samplingMethods[i % len(samplingMethods)])
            # Set consistent bounds for the charts
            plt.ylim(ymax = 1.1, ymin = 0)
            # Create a variable to iterate the different sources
            z = 0
            # Finally import the data! 
            for fileHeader in sourcesToRead:
                csvFile =  pd.read_csv(fileHeader + samplingMethod, header=0, 
                              converters={'Alpha (%)': lambda x: float(x.replace('%', '')),
                                          'BFS:RND': lambda x: float(x.replace('%', '')),
                                          'Percentage Amount': lambda x: float(x.replace('%', '')),
                                          'Sample Correlation Alpha (%)': lambda x: float(x.replace('%', '')),
                                          'Population P/R Alpha': lambda x: float(x.replace('%', ''))
                                          })
                #Filter to the correct measure
                csvFile = csvFile[(csvFile['Measure'] == measure) & (csvFile['Sample P/R Alpha'] == precision)]
                # Debug print
                # print fileHeader + samplingMethod
                # Remove duplicates
                theFrame = pd.DataFrame(data=csvFile[columns])
                theFrame.drop_duplicates(cols=columns, take_last=True, inplace = True)
                # Group the data
                the_grouped = theFrame.groupby(grouping)
                # Generate the values needed for the plot
                data_m = np.array(the_grouped.mean()['Precision'], float)
                data_df = np.array(the_grouped.size()) - 1
                data_sd = np.array(the_grouped.std()['Precision'], float)
                
                # Debugging
                # print str(z) + ',' + str(len(graphs))
                # print graphs[z%len(graphs)]
                
                plot.errorbar(alpha, data_m, label=graphs[z%len(graphs)], yerr = ss.t.ppf(0.95, data_df)*data_sd, xerr=0)
                theFrame.to_csv("files/precision - " + measure + samplingMethod, header = True)
                # plot.plot(alpha, data_m, label=graphs[z%len(graphs)]) #, yerr = ss.t.ppf(0.95, data_df)*data_sd, xerr=0)
                z = z + 1
            
            plot.set_ylabel('Precision (Sample Alpha = ' + str(precision) + ')')
            plot.set_xlabel('Alpha (%)')
            i = i + 1
        plt.legend(loc='bottom')
    fig.savefig('images/precision - ' + measure + '.png') 
```

```python
#Only doing Correlation in this section
columns = ['Parent Node Count', 'Parent Edge Count', 'Sample Method Type', 'BFS:RND', 'Replica ID', 'Alpha (%)', 'Sample Node Count', 'Sample Edge Count', 'Measure', 'Sample Correlation Alpha (%)', 'Spearmans', 'Pearsons', 'Error']
grouping = ['Parent Node Count', 'Parent Edge Count', 'Sample Method Type', 'BFS:RND', 'Alpha (%)', 'Measure', 'Sample Correlation Alpha (%)']

sampleCorrOptions = [5, 10, 100]

print '\033[5m' + 'Pearson\'s Rank Correlation' + '\033[0m'

for measure in measuresTaken:        
    # Make it known what the measure is
    print measure
    i = 1
    fig = plt.figure(figsize=(40, 20))
    for sampleCorr in sampleCorrOptions:
        # sampleCorr is known as part of the y axis
        for samplingMethod in filesToRead:
            # Add the plot and the sampling method corresponding
            plot = fig.add_subplot(len(sampleCorrOptions), len(samplingMethods), i)
            plot.set_title(samplingMethods[i % len(samplingMethods)])
            # Set consistent bounds for the charts
            plt.ylim(ymax = 1.1, ymin = 0)
            # Create a variable to iterate the different sources
            z = 0
            # Finally import the data! 
            for fileHeader in sourcesToRead:
                csvFile =  pd.read_csv(fileHeader + samplingMethod, header=0, 
                              converters={'Alpha (%)': lambda x: float(x.replace('%', '')),
                                          'BFS:RND': lambda x: float(x.replace('%', '')),
                                          'Percentage Amount': lambda x: float(x.replace('%', '')),
                                          'Sample Correlation Alpha (%)': lambda x: float(x.replace('%', '')),
                                          'Population P/R Alpha': lambda x: float(x.replace('%', ''))
                                          })
                
                #Filter to the correct measure
                csvFile = csvFile[(csvFile['Measure'] == measure) & 
                                  (csvFile['Sample Correlation Alpha (%)'] == sampleCorr)]
                # Debug print
                # print fileHeader + samplingMethod
                # Remove duplicates
                theFrame = pd.DataFrame(data=csvFile[columns])
                theFrame.drop_duplicates(cols=columns, take_last=True, inplace = True)
                # Group the data
                the_grouped = theFrame.groupby(grouping)
                # Generate the values needed for the plot
                data_m = np.array(the_grouped.mean()['Pearsons'], float)
                data_df = np.array(the_grouped.size()) - 1
                data_sd = np.array(the_grouped.std()['Pearsons'], float)
                
                plot.errorbar(alpha, data_m, label=graphs[z%len(graphs)], yerr = ss.t.ppf(0.95, data_df)*data_sd, xerr=0)
                z = z + 1
            plot.set_ylabel('Pearson\'s Rank Correlation (Top ' + str(sampleCorr) + ' %)')
            plot.set_xlabel('Alpha (%)')
            i = i + 1
        plt.legend(loc='bottom')
    fig.savefig('images/pearsons - ' + measure + '.png')
```
