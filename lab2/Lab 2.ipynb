Construct a dataframe with two columns: `country_name`, `num_players`. Use `groupby`.

```python
merged = pd.merge(players, countries, on='country_id', sort=True, how='left')
merged.rename(columns={'name_y':'country_name'}, inplace=True)

df = pd.DataFrame()
df['num_players'] = merged.groupby('country_name').size()
df.reset_index(level=0, inplace=True)
df
```

Convert the above left outer join query (last question in the SQL assignment) into pandas equivalent. See http://pandas.pydata.org/pandas-docs/stable/merging.html for details on how to do outer joins in pandas.

select c.name, count(p.name)
              from countries c inner join ( 
              SELECT name, country_id FROM PLAYERS where extract(year from birthdate) = 1975) p
              on c.country_id = p.country_id
              group by c.name;

```python
fPlay = players[pd.to_datetime(players['birthdate']).apply(lambda x:x.year) == 1975][['country_id', 'player_id']]
mCols = pd.merge(countries, fPlay, on='country_id', how='outer')
df = mCols.groupby(['country_id', 'name']).count()[['player_id']]
df.reset_index(level=0, inplace=True)
df.reset_index(level=1, inplace=True)
df.rename(columns={'name':'country', 'player_id':'player_count'}, inplace=True)
df[['country', 'player_count']]
```
